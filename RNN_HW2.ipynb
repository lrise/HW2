```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import psutil
import os

# For text preprocessing
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# For model building
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GRU, Dense, Dropout, SpatialDropout1D
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import nltk
nltk.download('punkt_tab')

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# 1. Load and explore the data
print("Loading data...")
train_df = pd.read_csv('/content/train.csv')
test_df = pd.read_csv('/content/test.csv')

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# Display first few rows
print("\nFirst few rows of training data:")
train_df.head()

# Check for missing values
print("\nMissing values in train data:")
print(train_df.isnull().sum())

# Data distributions
print("\nTarget distribution in training data:")
print(train_df['target'].value_counts())
print(train_df['target'].value_counts(normalize=True) * 100)

# 2. Text Preprocessing
def preprocess_text(text):
    if isinstance(text, str):
        # Convert to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r'https?://\S+|www\.\S+', '', text)

        # Remove HTML tags
        text = re.sub(r'<.*?>', '', text)

        # Remove mentions and hashtags (but keep the hashtag text)
        text = re.sub(r'@\w+', '', text)
        text = re.sub(r'#(\w+)', r'\1', text)

        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))

        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text).strip()

        # Tokenize
        tokens = word_tokenize(text)

        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [word for word in tokens if word not in stop_words]

        # Lemmatize
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]

        # Rejoin tokens
        text = ' '.join(tokens)

        return text
    return ""

# Apply preprocessing to the text columns
print("Preprocessing text data...")
train_df['cleaned_text'] = train_df['text'].apply(preprocess_text)
test_df['cleaned_text'] = test_df['text'].apply(preprocess_text)

# Display some examples of cleaned text
print("\nExamples of cleaned text:")
for i in range(5):
    print(f"Original: {train_df['text'].iloc[i]}")
    print(f"Cleaned: {train_df['cleaned_text'].iloc[i]}")
    print(f"Target: {train_df['target'].iloc[i]}")
    print("---")

# 3. Tokenization and Sequence Generation
# Tokenize the text
MAX_VOCAB_SIZE = 10000
MAX_SEQUENCE_LENGTH = 100

tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token='<OOV>')
tokenizer.fit_on_texts(train_df['cleaned_text'])

# Convert text to sequences
train_sequences = tokenizer.texts_to_sequences(train_df['cleaned_text'])
test_sequences = tokenizer.texts_to_sequences(test_df['cleaned_text'])

# Pad sequences
X_train = pad_sequences(train_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')
X_test = pad_sequences(test_sequences, maxlen=MAX_SEQUENCE_LENGTH, padding='post', truncating='post')

# Target variable
y_train = train_df['target'].values

# Get vocabulary size
vocab_size = min(MAX_VOCAB_SIZE, len(tokenizer.word_index) + 1)
print(f"Vocabulary size: {vocab_size}")

# Split the data into training and validation sets
X_train_data, X_val_data, y_train_data, y_val_data = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"Training data shape: {X_train_data.shape}")
print(f"Validation data shape: {X_val_data.shape}")

# 4. Model Building - LSTM
def build_lstm_model():
    embedding_dim = 128

    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=MAX_SEQUENCE_LENGTH))
    model.add(SpatialDropout1D(0.2))
    model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

    return model

# 5. Model Building - GRU (implemented from scratch)
def build_gru_model():
    embedding_dim = 128

    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=MAX_SEQUENCE_LENGTH))
    model.add(SpatialDropout1D(0.2))
    model.add(GRU(128, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

    return model

# 6. Training and Evaluation
batch_size = 64
epochs = 10

# Callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Helper function to get memory usage
def get_memory_usage():
    return psutil.Process(os.getpid()).memory_info().rss / 1024 / 1024  # in MB

# Training function to measure time and memory
def train_model(model, model_name):
    print(f"\nTraining {model_name} model...")

    # Measure memory before training
    memory_before = get_memory_usage()
    print(f"Memory usage before training: {memory_before:.2f} MB")

    # Measure training time
    start_time = time.time()

    # Train the model
    history = model.fit(
        X_train_data, y_train_data,
        validation_data=(X_val_data, y_val_data),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stopping],
        verbose=1
    )

    # Calculate training time
    training_time = time.time() - start_time
    print(f"{model_name} training time: {training_time:.2f} seconds")

    # Measure memory after training
    memory_after = get_memory_usage()
    memory_usage = memory_after - memory_before
    print(f"Memory usage after training: {memory_after:.2f} MB")
    print(f"Memory increase during training: {memory_usage:.2f} MB")

    # Evaluate on validation set
    val_loss, val_acc = model.evaluate(X_val_data, y_val_data, verbose=0)
    print(f"{model_name} validation accuracy: {val_acc:.4f}")

    # Predictions on validation set
    y_pred = (model.predict(X_val_data) > 0.5).astype(int)

    # Print classification report
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_val_data, y_pred))

    # Calculate confusion matrix
    cm = confusion_matrix(y_val_data, y_pred)

    # Plot training history
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'{model_name} Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'{model_name} Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.savefig(f'{model_name}_training_history.png')
    plt.show()

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{model_name} Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.savefig(f'{model_name}_confusion_matrix.png')
    plt.show()

    return {
        'model': model,
        'history': history,
        'training_time': training_time,
        'memory_usage': memory_usage,
        'val_accuracy': val_acc
    }

# Train LSTM model
lstm_model = build_lstm_model()
lstm_model.summary()
lstm_results = train_model(lstm_model, "LSTM")

# Train GRU model
gru_model = build_gru_model()
gru_model.summary()
gru_results = train_model(gru_model, "GRU")

# 7. Model Comparison
print("\n===== Model Comparison =====")
print(f"LSTM training time: {lstm_results['training_time']:.2f} seconds")
print(f"GRU training time: {gru_results['training_time']:.2f} seconds")
print(f"LSTM memory usage: {lstm_results['memory_usage']:.2f} MB")
print(f"GRU memory usage: {gru_results['memory_usage']:.2f} MB")
print(f"LSTM validation accuracy: {lstm_results['val_accuracy']:.4f}")
print(f"GRU validation accuracy: {gru_results['val_accuracy']:.4f}")

# Visualize comparison
metrics = ['Training Time (s)', 'Memory Usage (MB)', 'Validation Accuracy']
lstm_values = [lstm_results['training_time'], lstm_results['memory_usage'], lstm_results['val_accuracy']]
gru_values = [gru_results['training_time'], gru_results['memory_usage'], gru_results['val_accuracy']]

# Create a comparative bar chart
plt.figure(figsize=(12, 6))
x = np.arange(len(metrics))
width = 0.35

plt.bar(x - width/2, lstm_values, width, label='LSTM')
plt.bar(x + width/2, gru_values, width, label='GRU')

plt.title('LSTM vs GRU Performance Comparison')
plt.xticks(x, metrics)
plt.ylabel('Value')
plt.legend()

plt.tight_layout()
plt.savefig('model_comparison.png')
plt.show()

# 8. Making Predictions on Test Data
def generate_predictions(model, model_name):
    print(f"\nGenerating predictions with {model_name} model...")
    predictions = (model.predict(X_test) > 0.5).astype(int)
    return predictions.flatten()

# Generate predictions
lstm_predictions = generate_predictions(lstm_model, "LSTM")
gru_predictions = generate_predictions(gru_model, "GRU")

# 9. Creating CSV output
output_df = pd.DataFrame({
    'id': test_df['id'],
    'LSTM': lstm_predictions,
    'GRU': gru_predictions
})

print("\nPrediction output (first few rows):")
print(output_df.head())

# Save to CSV
output_df.to_csv('/content/predictions.csv', index=False)
print("Predictions saved to 'predictions.csv'")

# 10. Summary

print("\n===== Summary =====")
print(f"LSTM Model:")
print(f"- Training time: {lstm_results['training_time']:.2f} seconds")
print(f"- Memory usage: {lstm_results['memory_usage']:.2f} MB")
print(f"- Validation accuracy: {lstm_results['val_accuracy']:.4f}")

print(f"\nGRU Model:")
print(f"- Training time: {gru_results['training_time']:.2f} seconds")
print(f"- Memory usage: {gru_results['memory_usage']:.2f} MB")
print(f"- Validation accuracy: {gru_results['val_accuracy']:.4f}")

print("\nModel Differences:")
time_diff = gru_results['training_time'] - lstm_results['training_time']
memory_diff = gru_results['memory_usage'] - lstm_results['memory_usage']
accuracy_diff = gru_results['val_accuracy'] - lstm_results['val_accuracy']

print(f"- Training time: GRU is {abs(time_diff):.2f} seconds {'faster' if time_diff < 0 else 'slower'} than LSTM")
print(f"- Memory usage: GRU uses {abs(memory_diff):.2f} MB {'less' if memory_diff < 0 else 'more'} memory than LSTM")
print(f"- Validation accuracy: GRU is {abs(accuracy_diff)*100:.2f}% {'better' if accuracy_diff > 0 else 'worse'} than LSTM")

print("\nConclusion:")
if gru_results['val_accuracy'] > lstm_results['val_accuracy']:
    better_model = "GRU"
elif lstm_results['val_accuracy'] > gru_results['val_accuracy']:
    better_model = "LSTM"
else:
    better_model = "Both models performed equally"

print(f"The {better_model} model performed better in terms of prediction accuracy.")
```
